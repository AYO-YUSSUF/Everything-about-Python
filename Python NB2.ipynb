#SETS: is a collection which is unordered and unindexed. In python sets are written with curly brackets {   }

```python
#creating set
myset={"apple","banana","cherry"}
print(myset)
```

```python
#accessing items from the set
myset={"apple","banana","cherry"}
for i in myset:
  print(i)
```

#adding items to set by add() and update()
#BY ADD() YOU CAN ADD ONLY SINGLE ITEM
#BY UPDATE() YOU CAN ADD ONLY MUTIPLE ITEM

```python
#adding items to set by add() and update()
#BY ADD() YOU CAN ADD ONLY SINGLE ITEM
#BY UPDATE() YOU CAN ADD ONLY MUTIPLE ITEM
myset={"apple","banana","cherry"}
myset.add("orange")
print(myset)
```

```python
myset={"apple","banana","cherry"}
myset.update(["orange","mango","melon"])
print(myset)
```

```python
#len function
myset={"apple","banana","cherry"}
print(len(myset))
```

```python
#remove item from set
myset={"apple","banana","cherry"}
myset.remove("banana")
print(myset)
```

```python
myset={"apple","banana","cherry"}
myset.discard("banana")
myset.discard("sgh")

print(myset)
```

```python
#clear all items
myset={"apple","banana","cherry"}
myset.clear()
print(myset)
```

```python
#joining 2 sets - union()
set1={"a","d","g","h"}
set2={1,2,3}
set3=set1.union(set2)
print(set3)
```

```python
set1={"a","d","g","h"}
set2={1,2,3}
set1.update(set2)
print(set1)
```

#VARIABLES

##* LOCAL: the variables created inside of the functions are called local variables.
##* GLOBAL: the variables created outside of the functions are called local variables.

```python
global_var = 20  #global variable

def func():
  local_var=10   #created inside the function (local variable)
  print(local_var)
  print(global_var)
func()
```

#**YOU HAVE TO ALWAYS CALL THE FUNCTION AFTER PRINT COMMAND TO DISPLAY THE OUTPUT.**

```python
def cool():
  #global xy=200 #invalid syntax
  global xy
  xy = 200 #global variable
  print(xy)
cool()
```

#ARGUMENTS

* POSITIONAL ARGUMENTS
* KEYWORD ARGUMENTS

```python
def func(i,j):
  print(i,j)
func(646,82) #positional arguments
```

```python
def func(i,j):
  print(i,j)
func(j=646,i=82) #keyword arguments {order or position does not matter, we just give the command}
```

```python
def func(i,j=90):
  print(i,j)

func(100,200)
```

```python
#keyword arguments
def greetings(name,greetmsg):
  print(greetmsg+"   "+name)

greetings(name='Heemali', greetmsg="Hey!")
greetings(greetmsg="Hey!",name='Heems')
```

#**YOU CAN'T GIVE KEYWORD ARGUMENTS BEFORE POSITIONAL ARGUMENTS, AS SHOWN IN THE 10TH LINE OF THE CODE**

```python
def my_func(a,b,c):
  print(a,b,c)

my_func(10,20,30) #positional arguments
my_func(a=10,b=20,c=30) #keywords arguments
my_func(b=20,a=10,c=30) #keywords arguments

my_func(10,20,c=30)
my_func(10,b=20,c=30)
#my_func(10,b=20,30) #this is wrong as positional argument must appear before any keyword argument
```

#LOOPING STATEMENTS : WHILE LOOP: FOR LOOP


#While writing looping statements you have take into consideration three steps:

1.   Initialization
2.   Incrementation
3.   Condition



Printing numbers from 1-10 in ascending order

```python
i=1 #inilitilization
while i<=10: 
  print(i)
  i=i+1
print("Completed")
```

Printing numbers from 1-10 in descending order

```python
i=10
while i>=1:
  print(i)
  i=i-1
print("Completed")
```

#FOR LOOP

```python
 #printing 1-10 using for loop
 for i in range(1,11):
   print(i)
```

```python
for i in range(10,0,-1):
  print(i)
```

#JUMPING STATEMENTS: CONTINUE, BREAK

```python
for i in range(1,10):
  if i==5:
    break
  print(i)
print("program exited.")
```

```python
for i in range(1,10):
  if i==5:
    continue
  print(i)
print("program exited.")
```
